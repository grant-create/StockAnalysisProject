```python

import click
import robin_stocks as rh
import yfinance as yf
import pandas as pd
import pandas_datareader.data as web
import datetime
from statistics import mean
from datetime import date
from IPython.display import clear_output
import math

previousclose = []
lastclose = []
buy_on_open = []
price_for_buy = []
stock_open = []
stock_high = []
stock_low =[]
buy_sell_hold = []
position_held = []
option_price = []
twentyavg = []
c = 0
bolup = []
boldown = []
rolling_list = []



today = date.today()
endyear = today.strftime('%Y')
endmonth = today.strftime('%m')
endday = today.strftime('%d')

start = datetime.datetime(2020,11,1)
end = datetime.datetime(int(endyear),int(endmonth),int(endday))


#to get list of S&P500 stocks
payload=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
first_table = payload[0]
second_table = payload[1]

df = first_table
symbols = df['Symbol'].values.tolist()

diff = []
diff_sum = 0
bol = []


# The loop to get the close information for each column
for stock in symbols:
    c +=1
    try:
        stock_close =web.DataReader(stock,'yahoo',start,end)
        #Standard devs differences/sum
        lt= stock_close['Close'][-21:]
           
        twenty_avg = mean(lt)
        lc= stock_close['Close'][-1]
        
        if lc < twenty_avg and lc < 50:
            
        
            rolling_list.append(stock)
            
            buy_on_open.append(stock)

            last_twenty= stock_close['Close'][-21:].tolist()
            diff = []    
            twenty_avg = mean(last_twenty)
            for num in last_twenty:
                v = (num - twenty_avg)**2
                diff.append(v)
            diff_sum = sum(diff)

            std = math.sqrt(diff_sum/21)
            bol.append(std*2)
            bol_up = twenty_avg + (std*2)
            bol_down = twenty_avg - (std*2)

            stock_today_open= stock_close['Open'][-1].tolist()
            stock_today_high= stock_close['High'][-1].tolist()
            stock_today_low= stock_close['Low'][-1].tolist()
            last_close= stock_close['Close'][-1].tolist()

            stock_open.append(round(stock_today_open,2))
            stock_high.append(round(stock_today_high,2))
            stock_low.append(round(stock_today_low,2))
            lastclose.append(round(last_close,2))
            twentyavg.append(round(twenty_avg,2))
            bolup.append(round(bol_up,2))
            boldown.append(round(bol_down,2))
            
        else:
            pass
        
       
            

        clear_output()
        print(stock)
        print("{} of 500".format(c))

    except KeyError:
        
        pass
    
    
    
    
    


#"Last Close": lastclose, "Open": stock_open, "High": stock_high, "Low": stock_low, 

r = {"Rolling": twentyavg, "Last Close": lastclose, "Lower Band": boldown, "Upper Band": bolup}


rdf = pd.DataFrame(data = r, index = buy_on_open)


pd.set_option("display.max_rows", None, "display.max_columns", None)
#databyprice = df.sort_values(by=['Price'])
rdf




rdf
```
